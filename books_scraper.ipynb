```python
import requests
from bs4 import BeautifulSoup as bs
from tqdm import tqdm

```

```python
data=[]
for i in tqdm(range(1,51)):
    url=f'https://books.toscrape.com/catalogue/page-{i}.html'
    response=requests.get(url)
    soup=bs(response.text,'html.parser')
    for book in soup.find_all('li',class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
        title=book.find('img').get('alt')
        link='https://books.toscrape.com/catalogue/'+book.find('a').get('href')
        img_link='https://books.toscrape.com/'+book.find('img').get('src')[2:]
        rating=book.find('p').get('class')[-1]
        price=book.find('p', class_='price_color').text[1:]
        availability=book.find('p',class_='instock availability').text.strip()
        data.append([title,link,img_link,rating,price])
```
